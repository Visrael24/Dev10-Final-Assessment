```python
import pandas as pd

```

```python
pop2009 = pd.read_csv('../Becca_workspace/data/pop200to2009.csv', encoding='ISO-8859-1')
pop2019 = pd.read_csv('../Becca_workspace/data/pop2010to2019.csv', encoding='ISO-8859-1')
pop2021 = pd.read_csv('../Becca_workspace/data/pop2020to2021.csv', encoding='ISO-8859-1')
```

###TRANSFORMATION

##POP2006-2009

```python
pop2009[['NewName', 'city']] = pop2009.NAME.str.split(" city", expand = True)
```

```python
pop2009[['CityName', 'town']] = pop2009.NewName.str.split(" town", expand = True)
```

```python
pop2009.drop(pop2009[pop2009['SUMLEV'] != 162].index, inplace = True)
```

```python
pop2009.drop(['PLACE', 'SUMLEV', 'COUNTY', 'COUSUB', 'NAME', 'POPCENSUS_2000', 'POPBASE_2000', 'POP_2000', 'POP_2001', 'POP_2002', 'POP_2003', 'POP_2004', 'POP_2005', 'NewName', 'city', 'town'], axis=1, inplace = True)
```

```python
pop2009.to_csv('../Becca_workspace/data/cleaned_pop2009.csv', index=False)
```

#POP2010-2019

```python
pop2019[['NewName', 'city']] = pop2019.NAME.str.split(" city", expand = True)
```

```python
pop2019[['CityName', 'town']] = pop2019.NewName.str.split(" town", expand = True)
```

```python
pop2019.drop(pop2019[pop2019['SUMLEV'] != 162].index, inplace = True)
```

```python
pop2019.drop(['PLACE', 'STNAME', 'SUMLEV', 'COUNTY', 'COUSUB', 'CONCIT', 'PRIMGEO_FLAG', 'FUNCSTAT', 'NAME', 'CENSUS2010POP', 'ESTIMATESBASE2010', 'POPESTIMATE2017', 'POPESTIMATE2018', 'POPESTIMATE2019', 'NewName', 'city', 'town'], axis=1, inplace = True)
```

```python
pop2019.to_csv('../Becca_workspace/data/cleaned_pop2016.csv', index=False)
```
