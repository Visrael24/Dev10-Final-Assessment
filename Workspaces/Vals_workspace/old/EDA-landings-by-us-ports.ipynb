```python
import pandas as pd
import matplotlib.pyplot as plt
pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv('../../Data/Fish/LANDINGS BY TOP US PORTS.csv')
```

```python
df = df.rename(columns=df.iloc[0])
df = df.iloc[1:]
df.reset_index(drop=True, inplace=True)
```

```python
df.columns = df.columns.str.strip()
```

```python
df
```

```python
df['Millions of Dollars'] = df['Millions of Dollars'].astype(float)
df['Millions of Pounds'] = df['Millions of Pounds'].astype(float)
df['Year'] = df['Year'].astype(int)
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['Millions of Pounds'], df['Millions of Dollars'])
plt.xlabel('Millions of Pounds')
plt.ylabel('Millions of Dollars')
plt.title('Relationship between Millions of Pounds and Millions of Dollars')
plt.show()

```

```python
plt.figure(figsize=(10, 6))
df.groupby('Year')['Millions of Pounds'].sum().plot(label='Millions of Pounds')
df.groupby('Year')['Millions of Dollars'].sum().plot(label='Millions of Dollars')
plt.xlabel('Year')
plt.ylabel('Value')
plt.title('Change in Millions of Pounds and Millions of Dollars over Years')
plt.legend()
plt.grid(True)
plt.show()
```
